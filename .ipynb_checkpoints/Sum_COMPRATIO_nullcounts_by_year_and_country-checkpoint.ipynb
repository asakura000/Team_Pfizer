```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

sns.set()

%matplotlib inline

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv('Pfizer_2021_v2.csv', converters={'POS_LEVEL1': lambda x: str(x), 'POS_LEVEL2':lambda x: str(x),
                                          'POS_LEVEL3': lambda x: str(x), 'POS_LEVEL4': lambda x: str(x),
                                          'POS_LEVEL5': lambda x: str(x)})
```

```python
df.head()
```

```python
def create_year(date_str):
    year = date_str[-4:]
    return int(year)
```

```python
df['Year'] = df['Month_Year'].apply(create_year)
```

```python
df.head()
```

```python
#df.isnull().sum()
```

```python
comparatio_null = df[df['COMPARATIO'].isnull()]
```

```python
comparatio_null.shape
```

```python
comparatio_null.head()
```

```python
comparatio_null['Year'].unique()
```

```python
filt_2017 = comparatio_null['Year'] == 2017
filt_2018 = comparatio_null['Year'] == 2018
filt_2019 = comparatio_null['Year'] == 2019
filt_2020 = comparatio_null['Year'] == 2020
filt_2021 = comparatio_null['Year'] == 2021
```

```python
crn_2017 = comparatio_null[filt_2017]
```

```python
crn_2017.head()
```

```python
gb_17 = crn_2017.groupby('COUNTRY')['RandomID'].count()

count_17 = pd.DataFrame(gb_17)

count_17 = count_17.reset_index()

count_17 = count_17.rename(columns={'RandomID': 'Sum'})
#count_17.head()
count_17.to_csv('compratio_missing_2017.csv', index=False)
```

```python
count_17.shape
```
