### We need to replace the following:
1. Viet Nam -> Vietnam
2. Brunei Darussalam -> Brunei
3. Ivory Coast -> Cote d'Ivoire

```python
def fix_country_names(country_name):
    if country_name == 'Viet Nam':
        country_name = 'Vietnam'
    elif country_name == 'Brunei Darussalam':
        country_name = 'Brunei'
    elif country_name == 'Ivory Coast':
        country_name = "Cote d'Ivoire"
    return country_name
```

```python
df['COUNTRY'] = df['COUNTRY'].apply(fix_country_names)
```

```python
def create_year(date_str):
    year = date_str[-4:]
    return int(year)
```

# **Change POS_LEVEL* columns to be string data types**

```python
# POS_LEVEL1 column has no null values, so handle it later:
pos_lv_cols = ['POS_LEVEL2', 'POS_LEVEL3', 'POS_LEVEL4', 'POS_LEVEL5']

# convert all values non-null values in POS_LEVEL2 - 5 to become strings
for col in pos_lv_cols:
    df[col] = df[col].fillna(-1)
    df[col] = df[col].astype(int)
    df[col] = df[col].astype(str)
    df[col] = df[col].replace('-1', np.nan)
    
# convert POS_LEVEL1 to become strings
df['POS_LEVEL1'] = df['POS_LEVEL1'].astype(str)

# function to add leading zeros to the strings 
def add_zeros(a_str):
    zeros = ''
    diff = 0
    str_l = len(a_str)
    diff = 8 - str_l
    if diff > 0:
        for i in range(diff):
            zeros += '0'
    final_str = zeros + a_str
    return final_str

# redefine the list of columns bc this next operation will be done on all 5 columns
pos_lv_cols = ['POS_LEVEL1', 'POS_LEVEL2', 'POS_LEVEL3', 'POS_LEVEL4', 'POS_LEVEL5']

# apply the leading zeros function to all 5 columns
for col in pos_lv_cols:
    df[col] = df[col].fillna('some_null_string')
    df[col] = df[col].apply(lambda x: add_zeros(x))
    df[col] = df[col].replace('some_null_string', np.nan)
```
